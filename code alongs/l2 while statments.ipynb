## Example milk 

````python
# 1L milk 
# 1500000 bacterias
# Bactara increases with 50% each hour
# milk get soured after 10000000
#how many hours does it take til milk sours

import matplotlib.pyplot as plt

bacterias = 1.5e6 
factor = 1.5
hours = 0
sour = 1e7


bacterias_list = [bacterias]



print(bacterias_list)





while bacterias < sour: 
    bacterias *= factor
    hours +=1
    bacterias_list.append(bacterias) # adds element into the list 
print(bacterias_list)
print(f"takes {hours} for milk to get soured")

hours_list = range(0, hours+1)

plt.plot(hours_list, bacterias_list, 'o-')

plt.grid()
plt.xlabel ("hours")
plt.ylabel("number of bacteria")
plt.title("Bacterias in milk")

````
