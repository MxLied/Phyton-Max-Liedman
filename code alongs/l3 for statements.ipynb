## Random module

````python
import random as rnd 

dice = rnd.randint(1,6)
number_six = 0
numberrolls=1000000
for i in range (numberrolls):
    dice = rnd.randint(1,6)
    if dice == 6:
        number_six += 1
print(f"{number_six}")
print(f"proportion of sixes{number_six/numberrolls}")
````

````python
import matplotlib.pyplot as plt
#creates a empty list
pasta_per_year = []

# loops 140 years - because 2100-1960
for year in range (140):
    pasta_consump = .791*math.exp(.0526*year)
    #append - adds an element to the list
    pasta_per_year.append(pasta_consump)
pasta_per_year

plt.plot(range(1960,2100), pasta_per_year)
plt.xlabel("År")
plt.ylabel("pasta konsumpution i kg")
plt.title("Pastakonsumption per år i kg från 1960")

print("väldigt mycket pasta")
````
