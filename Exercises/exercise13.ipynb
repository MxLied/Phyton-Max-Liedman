```python
# TODO: class video


class Video: 
    def __init__(self, title: str,genre:str,rating: float):
        self.title = title
        self.genre = genre
        self.rating = rating

    @property
    def title(self):
        return self._title


    @title.setter
    def title(self, title_str: str):

        if type(title_str) != str:
            raise TypeError(f"Value has to be string")
        if title_str.strip() == "":
            raise ValueError(" xd")
        title = title_str.capitalize

        self._title = title_str



    @property 
    def genre(self):
        return self.genre

    @genre.setter
    def genre(self, genre_str :str):
        genre_lst = ["action", "comedy", "drama", "fantasy", "horror", "mystery", "romance", "thriller"]
        if genre_str.lower() not in genre_lst:
            raise ValueError(f"{genre_lst} is not a valid genre")
        genre_str = genre_str.capitalize()

        self._genre = genre_str


    @property
    def rating(self):
        return self.rating


    @rating.setter 
    def rating(self, float_rating):
        if not isinstance(float_rating, (int, float)):
            raise TypeError("rating must be a number")

        if not 0 < float_rating < 10:
            raise ValueError("rating must be 0 10")

        self._rating = float_rating


    def info(self) -> str:
        if type(self) == Video:
            return f"was {self._genre} {self._title} {self._rating}"

##########################################################################################
class TV_serie(Video):
    def __init__(self, title: str, genre: str, rating: float, episodes: int):
        super().__init__(title, genre, rating)
        self.episodes = episodes
    """"class for tv series """
    
    @property
    def episodes(self):
        return self._episodes


    @episodes.setter 
    def episodes(self, int_episode: int):
       
        if not isinstance(int_episode, (int, float)):
            raise TypeError("rating must be a number")

        if not 0 < int_episode:
            raise ValueError("rating must be more then 0")

        self._episode = int_episode 

    def info(self) -> str:
        if type(self) == TV_serie:
            return f"was {self._genre} {self._title} {self._rating} {self._episode}"

pokemon = TV_serie("Pokemon", "fantasy", 3.1, 625)

print(pokemon.info())
```
